# Data Cleaning  

```python
# first we have to deal with the null values in columns
# first checking the % of the missing values
amazon = amazon.drop(['reviews.didPurchase'],axis = 1)
amazon = amazon.drop(['reviews.id'],axis = 1)

# dropping the columns since its almost all the values are null and it doesnot have any relevance to our analysis
```

```python
# check the % of missing values of reviews.didpurchase column
(12246/28332)*100
```

```python
# dropping just all the null values
amazon = amazon[~amazon['reviews.doRecommend'].isnull()]
amazon = amazon[~amazon['reviews.numHelpful'].isnull()]
```

```python
amazon.isnull().sum()
```

All the columns have no null values!!!

```python
# changing the columns into appropiate datatype
```

```python
amazon.dtypes
```

```python
# Date Time

amazon['dateAdded'] = pd.to_datetime(amazon['dateAdded'])
amazon['dateUpdated'] = pd.to_datetime(amazon['dateUpdated'])
```

```python
# creating the new columns
amazon['dateAdded_years'] = amazon['dateAdded'].dt.year
amazon['dateAdded_month'] = amazon['dateAdded'].dt.month

# Now, dropping the columns
amazon = amazon.drop(['dateAdded'],axis = 1)
```

```python
# creating the new columns
amazon['dateUpdated_years'] = amazon['dateUpdated'].dt.year
amazon['dateUpdated_month'] = amazon['dateUpdated'].dt.month

# Now, dropping the columns

amazon = amazon.drop(['dateUpdated'],axis = 1)
```

```python
# changing the daTA TYpe
amazon['reviews.date_year'] = pd.to_datetime(amazon['reviews.date']).dt.year
amazon['reviews.date_month'] = pd.to_datetime(amazon['reviews.date']).dt.month
```

```python
amazon = amazon.drop(['reviews.date'],axis = 1)
```

```python
# dropping the columns
var = ['asins','keys','manufacturerNumber','id']
amazon = amazon.drop(var,axis =1)
```

```python
amazon.head()
```

```python
amazon['reviews.dateSeen'] = amazon['reviews.dateSeen'].astype('str')
```

```python
amazon = amazon.drop(['imageURLs'],axis = 1)
amazon = amazon.drop(['sourceURLs'],axis = 1)
```

# Data Cleaning 

```python
# dropping the columns since most of the values were null
amazon_1 = amazon_1.drop(["reviews.dateAdded",'reviews.id'],axis = 1)
```

```python
# dropping the null values of "reviews.title" columns
amazon_1['reviews.title'] = amazon_1['reviews.title'].dropna(how = 'any',inplace = True)
```

```python
# changing into apporpiate datatype
amazon_1['dateAdded_year'] = pd.to_datetime(amazon_1['dateAdded']).dt.year
amazon_1['dateAdded_month'] = pd.to_datetime(amazon_1['dateAdded']).dt.month
amazon_1['reviews.date_year'] = pd.to_datetime(amazon_1['reviews.date']).dt.year
amazon_1['reviews.date_month'] = pd.to_datetime(amazon_1['reviews.date']).dt.month

# dropping the columns
amazon_1 = amazon_1.drop(["dateAdded",'reviews.date'],axis = 1)
```

```python
# data visualisation 
plt.figure(figsize=[6,12])
amazon_1.groupby(['reviews.rating'])['dateAdded_year'].sum().plot.bar()
plt.title('rating per years')
plt.xlabel('ratings'),plt.ylabel('years')
plt.show()
```

```python
# line plot
plt.figure(figsize = [8,6])
plt.style.use('ggplot')
amazon_1.groupby(['dateAdded_month'])['reviews.rating'].mean().plot()
plt.title('trends over months of ratings',fontdict = {'fontsize' : 25,'fontweight':23,})
plt.ylabel('Ratings')
plt.legend('ratings')
plt.grid('True')

plt.show()
```

```python
plt.figure(figsize = [7,7])
plt.style.use('fivethirtyeight')
plt.grid('True')
plt.hist(amazon_1['primaryCategories'],color = 'cyan',edgecolor ='black',align = 'left' )
plt.xticks(rotation = 90)
plt.title('frequency of Primary Categories', fontdict = {'fontsize':19,'fontweight':12 })
plt.show()
```

```python
# bar
amazon_1.groupby(['primaryCategories'])['reviews.numHelpful'].mean().hist()
plt.xticks(rotation = 90)
plt.title('reviews helpful for Categories')
plt.show()
```
